```python
import sys
#change this to directory with flspectrum.py
sys.path.insert(1, "C:\\Users\\amgib\\OneDrive\\Documents\\Flourescence Data")

import pandas as pd 
import matplotlib.pyplot as plt
import os
from scipy.signal import find_peaks, argrelmax
from scipy.signal import savgol_filter
from flspectrum import FlSpectrum

path ="C:\\Users\\amgib\\OneDrive\\Documents\\Raw Spectra Data\\2023\\112923\\AS\\"
```

```python
name = "RH30.asc"
bgname = "BG30.asc"
RH = 30

#import data
tempspectrum = FlSpectrum(path,name,bgname,RH)

#calibrate x-axis, adjust prominence to be about half the difference between baseline and 532 peak
prominence = 2500
dataoffset = 0
bgoffset = 0

tempspectrum.calibrate(prominence,dataoffset,bgoffset)

#smooth and subtract
smoothparam = 301
tempspectrum.bgSubtractAndSmooth(smoothparam) 

#fit guassian
p0=[650.0,0.01,7000.0]
tempspectrum.fitGuassian(p0)

#these are the two variables
print(tempspectrum.smoothmax)
print(tempspectrum.gausscenter)

print(tempspectrum.smoothmaxuncalib)
print(tempspectrum.gausscenteruncalib)

#NAME IT HERE
AS_RH30_112923 = tempspectrum
```
